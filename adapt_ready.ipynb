2. Given an unsorted array of integers, find the length of the longest continuous increasing subsequence (subarray).

```python
def long_cont(a):
  a.sort()
  count = 1
  longest = []
  for i in a:
    if i + 1 in a:
      count += 1
    else:
      longest.append(count)
      count = 1
  return max(longest)

print(long_cont([1,3,5,4,7]))
print(long_cont([2,2,2,2,2]))
```

3. Given a list of non negative integers, arrange them such that they form the largest number.

```python
from itertools import permutations
def greatest_no(l):
  large = []
  for i in permutations(l,len(l)):
    large.append(''.join(map(str,i)))
  return max(large)

print(greatest_no([10,2]))
print(greatest_no([3,30,34,5,9]))
```

4. Store all the"servlet-name", and "servlet-class" to a csv file from the attached sample_json.json file using Python.

```python
import json

f = open("DT A1 sample_json (1) (1) (3).json")
data = json.load(f)
data
```

```python
d = data['web-app']['servlet']
name = []
classes = []
for i in d:
    #print(i['servlet-name'])
    name.append(i['servlet-name'])
    #print(i['servlet-class'])
    classes.append(i['servlet-class'])

print(name)
print(classes)
```

```python
import pandas as pd
table = list(zip(name,classes))
df = pd.DataFrame(table)
df.columns = ["servlet-name","servlet-classes"]
df
```

```python
df.to_csv("extracted_file.csv",index=False)
```
